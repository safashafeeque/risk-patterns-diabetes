```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("diabetes.csv")

# Replace invalid zeros
cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
data[cols] = data[cols].replace(0, np.nan)

# Fill missing values
data.fillna(data.median(), inplace=True)

# Group by Outcome and calculate mean
grouped = data.groupby("Outcome").mean()

print(grouped)
```

```python
import matplotlib.pyplot as plt

# Compare glucose means
grouped["Glucose"].plot(kind="bar")
plt.title("Average Glucose by Outcome")
plt.ylabel("Glucose")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Compare BMI means
grouped["BMI"].plot(kind="bar")
plt.title("Average BMI by Outcome")
plt.ylabel("BMI")
plt.show()
```
